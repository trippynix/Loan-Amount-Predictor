# IMPORTING NECESSARY LIBRARIES FIRST

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# LOADING THE DATASETS 

```python
train_data = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
test_data = pd.read_csv("test_Y3wMUE5_7gLdaTN.csv")
```

# DATA PRE-PROCESSING

```python
#CLEANING THE DATA AND ADDING IN THE MISSING VALUES
mean = train_data["LoanAmount"].mean()
train_data["LoanAmount"].fillna(mean, inplace=True)
mean = train_data["Credit_History"].mean()
train_data["Credit_History"].fillna(mean, inplace=True)
#USING MEAN WHERE THERE ARE A LOT OF UNIQUE INT VALUES AND MODE WHERE THERE ARE LIMITED UNIQUE VALUES
mode = train_data["Loan_Amount_Term"].mode()[0]
train_data["Loan_Amount_Term"].fillna(mode, inplace=True)
mode = train_data["Dependents"].mode()[0]
train_data["Dependents"].fillna(mode, inplace=True)
mode = train_data["Married"].mode()[0]
train_data["Married"].fillna(mode, inplace=True)
mode = train_data["Gender"].mode()[0]
train_data["Gender"].fillna(mode, inplace=True)
mode = train_data["Self_Employed"].mode()[0]
train_data["Self_Employed"].fillna(mode, inplace=True)
```

```python
train_data.isna().sum()  #CHECKING IF THERE IS ANY MISSING VALUE PRESENT
```

```python
train_data.drop("Loan_ID", axis=1, inplace=True)
test_data.drop("Loan_ID", axis=1, inplace=True)
#DROPPING LOAN ID AS EACH TAB HAD A UNIQUE VALUE WITH A LETTER WHICH CAUSED PROBLEM WHILE NORMALISING AND ENCODING THE VALUES
```

# ENCODING THE VALUES

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
cat_cols = ["Gender", "Married", "Education", "Self_Employed", "Property_Area","Loan_Status", "Dependents"]
```

```python
for col in cat_cols:
    le.fit(train_data[col])
    train_data[col] = le.transform(train_data[col])

cat_cols = ["Gender", "Married", "Education", "Self_Employed", "Property_Area", "Dependents"]
    
for col in cat_cols:
    le.fit(test_data[col])
    test_data[col] = le.transform(test_data[col])
```

```python
train_data #CHECKING IF ALL THE VALUES ARE CONVERTED INTO NUMERICAL VALUES
```

```python
test_data
```

```python
train_data.corr() #CHECKING THE HIGH CORRELATING FACTOR WITH LOAN AMOUNT
#IN THIS CASE APPLICANT INCOME WAS HIGH CORRELATING WITH LOAN AMOUNT
```

```python
#SPLITTING THE TRAIN AND TEST DATA INTO FEATURES(X_TRAIN, X_TEST) AND TARGET(Y_TRAIN, Y_TEST)
X_train = train_data.drop("ApplicantIncome", axis=1)
y_train = train_data["LoanAmount"]
X_test = train_data.drop("ApplicantIncome", axis=1)
y_test = train_data["LoanAmount"] 
```

# NORMALIZING THE VALUES

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()  #INITIALIZE THE SCALER
```

```python
scaler.fit(X_train) #FITTING THE SCALER ON THE TRAINING DATA
#TRANSFORMING THE TRAINING AND TESTING DATA
X_train = scaler.transform(X_train) 
X_test = scaler.transform(X_test)
```

# LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
```

```python
#TRAINING THE LINEAR REGRESSION MODEL USING THE TRAIN DATA
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_test #CHECKING HOW MUCH CORRECT THE MODEL PREDICTED
```

```python
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae) #CALCULATING AND PRINTING THE MAE AND MSE
```

# POLY REGRESSION MODEL

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
```

```python
#TRAINING THE POLY REGRESSION MODEL WITH THE TRAIN DATA
regressor = LinearRegression()
regressor.fit(X_train_poly, y_train)
```

```python
X_test_poly = poly.transform(X_test)
```

```python
y_pred = regressor.predict(X_test_poly)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = np.mean(np.abs((y_test, y_pred)))
```

```python
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae) #CALCULATING AND PRINTING THE MAE AND MSE
```
